```python
import numpy as np
import pandas as pd
# 数据合并(pd.concat)
left=pd.DataFrame([[1,2],[3,4],[5,6]],index=list('ace'),columns="语文 数学".split(" "))
left
right=pd.DataFrame([[7,8],[9,10],[11,12],[13,14]],index=list('bcde'),columns="英语 综合".split(" "))
right
pd.merge(left,right,left_index=True,right_index=True,how='outer')
```

```python
# join 按照索引进行连接
#按照索引合并，但要求没有重叠的咧
left.join(right,how='outer')  #两表不能出现重叠的列名
```

```python
# pd.concat
# 注意指定轴方向，默认axis=0
# join指定合并方式，默认为outer
# Series合并时查看行索引有无重复
# numpy  concatenate
arr1 = np.random.randint(0, 10, (3, 4))
arr2 = np.random.randint(0, 10, (3, 4))
print(arr1)
print(arr2)
print(np.concatenate([arr1, arr2]))
print(np.concatenate([arr1, arr2], axis=1))
```

```python
# 2. pd.concat
# 注意指定轴方向，默认axis=0
# join指定合并方式，默认为outer
# Series合并时查看行索引有无重复
df1 = pd.DataFrame(np.arange(6).reshape(3,2),index=list('abc'),columns=['one','two'])
df2 = pd.DataFrame(np.arange(4).reshape(2,2)+5,index=list('ac'),columns=['three','four'
])
pd.concat([df1,df2],sort=False) #默认外连接，axis=0


df1 = pd.DataFrame(np.arange(6).reshape(3,2),index=list('abc'),columns=['one','two'])   #当列名相同时上下合并
df2 = pd.DataFrame(np.arange(4).reshape(2,2)+5,index=list('ac'),columns=['one','two'
])
pd.concat([df1,df2],sort=False,axis=0) #默认外连接，axis=0
pd.concat([df1,df2],sort=False,axis=1) #外连接，axis=1
```

```python
# 重塑和轴向旋转
data=pd.DataFrame(np.arange(6).reshape(2,3),
index=pd.Index(["老王","小刘"],name="姓名"),
columns=pd.Index(["语文",'数学','英语'],name='科目'))
print(data)
# 1. stack
# 将列索引旋转为行索引，完成层级索引
# DataFrame->Series
d=data.stack()  #stack() 将列索引变为行索引
type(d)
# 2. unstack
# 将层级索引展开
# Series->DataFrame
d.unstack()  
```

```python
d.unstack(0) 
```

```python
s=pd.Series(np.arange(4),index=list('abcd'))
s1=pd.Series([4,5,6],index=list('cde'))
sc=pd.concat([s,s1],keys=["data1",'data2'])
sc
# type(sc)
sc.unstack()
#stack() 默认过滤缺失数据
sc.unstack().stack()
#dropna=False 不去过滤缺失数据
sc.unstack().stack(dropna=False)
```

```python
df=pd.DataFrame({'fruit':['apple','banana','orange','apple','banana'],
                'color':['red','yellow','yellow','cyan','cyan'],
                'price':[8.5,6.8,5.6,7.8,6.4]
                })
df
```

```python
#分组
g=df.groupby(by="fruit") #可迭代(是一个二元元组)
g
# for i in  g:
#     print(i)
for name,group in  g:
    print(name)  #输出组名
    print("-"*30)
    print(group)   #输出数据块
    print(type(group))
```

```python
# 选取任意数据块
dict(list(df.groupby(by="fruit")))['apple']
```

```python
# 聚合
# 根据水果来求水果平均值
g=df.groupby(by="fruit")['price'].mean()
g
# 语法糖
df['price'].groupby(df['fruit']).mean()
#as_index   当需要行索引是使用
df.groupby(by="fruit",as_index=False)['price'].mean()
```

```python
#计算每种水果的差值
#自定义聚合函数
def  diff(arr):
    #agg aggregate 
    return arr.max()-arr.min()
df.groupby('fruit')['price'].agg(diff)
```

```python
# 对多列进行分组
g=df.groupby(by=["fruit",'color'])['price'].mean()
g
print(type(g))  #Series
g1=df['price'].groupby([df['fruit'],df['color']]).mean()
print(type(g1))  #Series
#将Series对象转化为DataFrame对象
g=df.groupby(by=["fruit",'color'])[['price']].mean()
print(type(g))  #DataFrame
g1=df[['price']].groupby([df['fruit'],df['color']]).mean()
print(type(g1))  #DataFrame
```

```python
# 补充
people=pd.DataFrame(np.random.randn(5,5),
                   columns=list("abcde"),
                    index="Sam Make Tony Jim Nana".split(" "))

# 通过字典进行分组
m={"a":'red',"b":'red',"c":'blue',"d":'blue',"e":'red',"f":'origin'}
```

```python
people.groupby(m,axis=1).sum()
```

```python

# 通过Series进行分组
s=pd.Series(m)
people.groupby(s,axis=1).count()
```

```python
# 通过函数进行分组
people.groupby(len).sum()
```
