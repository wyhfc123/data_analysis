```python
"""
    填充(以某种颜色自动填充两条曲线的闭合区域)
        - plt.fill_between(x,sin_x,cos_x,sin_x<cos_x,color,alpha)
            - x                         x轴的水平坐标
            - sin_x                     下边界曲线上点的垂直坐标
            - cos_x                     上边界曲线上点的垂直坐标
            - sin_x < cos < x           填充条件，为True时填充
            - color                     填充颜色 <str>
            - alpha                     透明度   <float：0-1>

"""
import numpy as np
from matplotlib import pyplot as plt
#绘制两条曲线 sin_x = sin(x)   cos_x = cos(x/2)/2 [0-8π]

n=1000
x = np.linspace(0,8*np.pi,n)
sinx = np.sin(x)      #第一条线
cosx = np.cos(x/2)/2  #第二条线
plt.figure("Fill",facecolor="lightgray")
plt.title("Fill",fontdict={"fontsize":20})
plt.grid(linestyle=":")
plt.xlabel("x",fontdict={"fontsize":14})
plt.ylabel("y",fontdict={"fontsize":14})
plt.tick_params(labelsize=10)
plt.plot(x,sinx,color="dodgerblue",label = r"$y=sin(x)$")
plt.plot(x,cosx,color="orangered",label = r"$y=\frac{1}{2}cos(\frac{x}{2})$")
plt.fill_between(x,sinx,cosx,sinx>cosx,color="dodgerblue",alpha=0.3)
plt.fill_between(x,sinx,cosx,sinx<cosx,color="orangered",alpha=0.3)
plt.legend()
plt.show()
```

```python
"""
    np.meshgrid()  网格化函数，从坐标向量返回坐标矩阵(二维数组)
    等高线图(二维表达三维)
        - plt.contour(x,y,z,8,colors,linewidths)
            - x                  网格坐标矩阵的x坐标(2维数组)
            - y                  网格坐标矩阵的y坐标(2维数组)
            - z                  网格坐标矩阵的z坐标(2维数组)
            - 8                  把等高线绘制成8部分
            - colors             等高线的颜色
            - linewidths         线宽
        - plt.clabel(cutr,inline_spacing,fmt,fontsize)
            - cutr               contour函数返回值
            - inline_spacing     文本旁边间距
            - fmt                格式化
            - fontsize           字体大小


        - plt.contourf(x,y,z,8,cmap)
            - x                  网格坐标矩阵的x坐标(2维数组)
            - y                  网格坐标矩阵的y坐标(2维数组)
            - z                  网格坐标矩阵的z坐标(2维数组)
            - 8                  把等高线绘制成8部分
            - cmap               颜色映射

"""
import numpy as np
from matplotlib import pyplot as plt
n=1000
x,y = np.meshgrid(np.linspace(-3,3,n),np.linspace(-3,3,n)) #x,y直接组成坐标点矩阵
z = (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2) #z为通过每个坐标的x与y计算而得的高度值(模拟采集的海拔高度)
plt.figure("Coutour",facecolor="lightgray")
plt.title("Coutour",fontdict={"fontsize":20})
cutr = plt.contour(x,y,z,8,colors="black",linewidths=0.5)
#设置等高线上的高度标签文本
plt.clabel(cutr,inline_spacing=1,fmt="%.2f",fontsize=10)
plt.contourf(x,y,z,8,cmap="jet")
plt.show()
```

```python
"""
    三维散点图
        - ax3d.scatter()
            - x             x轴坐标数组
            - y             y轴坐标数组
            - z             z轴坐标数组
            - marker        点型
            - s             大小
            - zorder        图层序号
            - color         颜色
            - edgecolor     边缘颜色
            - facecolor     填充色
            - c             颜色值,根据cmap映射应用相应颜色
            - cmap          颜色映射
"""
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d   #必须导入

n = 300
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)
z = np.random.normal(0,1,n)
d = x**2 + y**2 + z**2
plt.figure("3D scatter",facecolor="lightgray")
ax3d = plt.gca(projection="3d")
ax3d.set_xlabel("x")
ax3d.set_ylabel("y")
ax3d.set_zlabel("z")

ax3d.scatter(x,y,z,s=70,marker="o",c=d,cmap="jet")
plt.show()
```

```python
"""
    简单动画
        - am.FuncAnimation(plt.gcf(),update,interval=10)
            - plt.gcf()         获取当前窗口
            - update            更新函数
            - interval          间隔时间(单位:毫秒)



"""

import matplotlib.animation as am
from matplotlib import pyplot as plt
import numpy as np
#自定义一种可以存放在ndarray里的类型，用于保存一个球
ball_type=np.dtype([
    ("position",float,2),           #位置(水平或垂直坐标)
    ("size",float,1),               #大小
    ("growth",float,1),             #生长速度(膨胀速度)
    ("color",float,4),              #颜色(红、绿、蓝和透明度)

])
#随机生成100个点对象
n =100
balls = np.zeros(100,dtype=ball_type)   #[([0., 0.], 0., 0., [0., 0., 0., 0.])*100]
#初始化balls数组每个字段的属性值
balls["position"] = np.random.uniform(0,1,(n,2))
# print(balls)
balls["size"] = np.random.uniform(50,70,n)
balls["growth"] = np.random.uniform(10,20,n)
balls["color"] = np.random.uniform(0,1,(n,4))

#画图
plt.figure("Animation",facecolor="lightgray")
plt.title("Animation",fontdict={"fontsize":20})
plt.xticks([])
plt.yticks([])
sc = plt.scatter(balls["position"][:,0],balls["position"][:,1],s = balls["size"],color=balls["color"])

#更新函数
def update(number):
    """
    :param number:  累加器
    :return:
    """
    #选择一个点
    index = number % 100
    #重新修改index位置元素的属性
    balls["position"][index] = np.random.uniform(0,1,(1,2))
    balls["size"][index]=  np.random.uniform(50,70,1)

    balls["size"] += balls["growth"]
    #重新绘制界面
    sc.set_sizes(balls["size"])  #更新大小
    sc.set_offsets(balls["position"])  #更新位置

anim = am.FuncAnimation(plt.gcf(),update,interval=10)
plt.show()
```
